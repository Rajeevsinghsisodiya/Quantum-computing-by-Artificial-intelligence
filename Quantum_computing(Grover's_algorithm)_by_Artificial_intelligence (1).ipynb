# Quantum Teleportation

Quantum teleportation is a fundamental quantum information processing protocol that allows the transfer of quantum information between two parties. Below is an example code for simulating quantum teleportation using Qiskit.

This below machine learning code demonstrates the basic steps of quantum teleportation creating an entangled Bell pair, preparing an arbitrary state, Alice's operations, measurement, and Bob's operations. The results are then displayed and visualized using a histogram.

```python
from qiskit import Aer, QuantumCircuit, transpile, assemble
from qiskit.visualization import plot_histogram

def create_bell_pair(qc, a, b):
    """Creates a Bell pair between qubits a & b"""
    qc.h(a)  # Apply Hadamard gate to qubit a
    qc.cx(a, b)  # Apply CNOT gate with a as control and b as target

def alice_gates(qc, psi, a):
    """Performs Bell measurement and applies gates on Alice's qubit"""
    qc.cx(psi, a)
    qc.h(psi)

def measure_and_send(qc, a, b):
    """Measures the qubits a & b and sends classical bits"""
    qc.measure(a, 0)
    qc.measure(b, 1)

def bob_gates(qc, qubit, crz, crx):
    """Applies gates on Bob's qubit based on classical bits"""
    qc.x(qubit).c_if(crx, 1)  # Apply X gate if crx is 1
    qc.z(qubit).c_if(crz, 1)  # Apply Z gate if crz is 1

# Create a quantum circuit
qc = QuantumCircuit(3, 3)

# Create an entangled Bell pair between qubits 1 and 2
create_bell_pair(qc, 1, 2)

# Prepare an arbitrary state on qubit 0
qc.h(0)
qc.rx(1.57, 0)  # Apply X-rotation to qubit 0 (pi/2)

# Teleportation protocol
alice_gates(qc, 0, 1)
measure_and_send(qc, 0, 1)
bob_gates(qc, 2, 0, 1)

# Simulate the circuit
simulator = Aer.get_backend('qasm_simulator')
compiled_circuit = transpile(qc, simulator)
result = simulator.run(assemble(compiled_circuit)).result()

# Display the circuit
print("Quantum Teleportation Circuit:")
print(qc)

# Display the measurement outcomes
counts = result.get_counts(qc)
print("\nMeasurement Outcomes:")
print(counts)

# Plot the histogram of measurement outcomes
plot_histogram(counts)
```

# Grover's algorithm

 Grover's algorithm, which is a quantum algorithm for so-called unstructured search problems that offers a quadratic improvement over classical algorithms. What this means is that Grover's algorithm requires a number of operations on the order of the square-root of the number of operations required to solve unstructured search classically  which is equivalent to saying that classical algorithms for unstructured search must have a cost at least on the order of the square of the cost of Grover's algorithm. Grover's algorithm, together with its extensions and underlying methodology, turn out to be broadly applicable, leading to a quadratic advantage for many interesting computational tasks that may not initially look like unstructured search problems on the surface.

# Grover's algorithm

Below machine learning code implements Grover's algorithm. The algorithm offers a quadratic speedup compared to classical algorithms for unstructured search problems.

1.**Quantum Circuit Functions**:

initialize_s(qc, qubits): Applies a Hadamard gate to create an equal superposition of states for the specified qubits. Also applies Z-gates to convert |0 states to |1 states.

oracle(qc, input_qubits, output_qubit): Applies a controlled-X (CX) gate (controlled NOT gate) as the oracle for the marked element. This marks the desired state.
diffusion(qc, qubits): Applies the diffusion operator, which involves Hadamard gates, Z-gates, controlled-Z (CZ) gates, and additional Hadamard gates.

2. **Grover's Algorithm Function (grover_algorithm)**:

Takes the number of qubits n and the marked element as input.
Creates a quantum circuit with n+1 qubits, where the additional qubit is used as an output qubit.
Initializes the qubits in an equal superposition using initialize_s.
Defines the number of Grover iterations based on the square root of the search space size for quadratic speedup.
Applies Grover iterations, consisting of the oracle and diffusion operations.
Measures the first n qubits.

3. **Main Code**:

Sets the number of qubits (num_qubits) and the marked element (marked_element) to be searched.
Calls grover_algorithm to create the Grover's algorithm circuit.
Visualizes the circuit using grover_circuit.draw(output='mpl').
Chooses the Aer simulator backend ('aer_simulator').
Runs the circuit on the simulator using simulator.run(grover_circuit).result().
Retrieves and plots the histogram of measurement outcomes using plot_histogram(counts).

4. **Simulation and Visualization**:

The circuit is simulated using the Aer simulator backend.
The measurement outcomes are obtained and plotted as a histogram, showing the probability distribution of the states.

The marked element is set to 6 in this example. The quantum circuit is designed to amplify the probability of measuring the marked element when executed on a quantum computer.

# Quantum Circuit Functions

```python
from qiskit import QuantumCircuit, Aer, transpile, assemble
from qiskit.visualization import plot_histogram

def initialize_s(qc, qubits):
    # Apply Hadamard gate to create superposition
    qc.h(qubits)

    # Apply Z-gates to convert |0 states to |1 states
    qc.z(qubits)

def oracle(qc, input_qubits, output_qubit):
    # Apply CX gate as the oracle for the marked element
    qc.cx(input_qubits, output_qubit)

def diffusion(qc, qubits):
    # Apply Hadamard gates
    qc.h(qubits)

    # Apply Z-gates
    qc.z(qubits)

    # Apply controlled-Z (CZ) gates
    qc.cz(qubits[:-1], qubits[-1])

    # Apply additional Hadamard gates
    qc.h(qubits)

# Create a quantum circuit
n = 3  # Number of qubits
qc = QuantumCircuit(n, n)

# Initialize superposition
initialize_s(qc, range(n))

# Apply Oracle (for example, marking the state |110)
oracle(qc, [0, 1], 2)

# Apply Diffusion operator
diffusion(qc, range(n))

# Measure the qubits
qc.measure(range(n), range(n))

# Visualize the circuit
qc.draw('mpl')

# Simulate the circuit
simulator = Aer.get_backend('qasm_simulator')
compiled_circuit = transpile(qc, simulator)
result = simulator.run(compiled_circuit).result()

# Plot the histogram of measurement results
counts = result.get_counts(qc)
plot_histogram(counts)
```

# Grover's Algorithm Function (grover_algorithm)

```python
from qiskit import QuantumCircuit, Aer, transpile, assemble
from qiskit.visualization import plot_histogram

def initialize_s(qc, qubits):
    # Apply Hadamard gate to create superposition
    qc.h(qubits)

    # Apply Z-gates to convert |0 states to |1 states
    qc.z(qubits)

def oracle(qc, input_qubits, output_qubit, marked_element):
    # Apply CX gate as the oracle for the marked element
    qc.cx(input_qubits, output_qubit)

    # Apply X gate to the output qubit if the marked element is 1
    if marked_element:
        qc.x(output_qubit)

def diffusion(qc, qubits):
    # Apply Hadamard gates
    qc.h(qubits)

    # Apply Z-gates
    qc.z(qubits)

    # Apply controlled-Z (CZ) gates
    qc.cz(qubits[:-1], qubits[-1])

    # Apply additional Hadamard gates
    qc.h(qubits)

def grover_algorithm(n, marked_element):
    # Create a quantum circuit with n+1 qubits
    qc = QuantumCircuit(n + 1, n)

    # Initialize superposition
    initialize_s(qc, range(n + 1))

    # Define the number of Grover iterations for quadratic speedup
    num_iterations = int(round(3.14/4 * (2 ** (n/2))))

    # Apply Grover iterations
    for _ in range(num_iterations):
        # Apply Oracle
        oracle(qc, range(n), n, marked_element)

        # Apply Diffusion operator
        diffusion(qc, range(n + 1))

    # Measure the first n qubits
    qc.measure(range(n), range(n))

    # Simulate the circuit
    simulator = Aer.get_backend('qasm_simulator')
    compiled_circuit = transpile(qc, simulator)
    result = simulator.run(compiled_circuit).result()

    # Plot the histogram of measurement results
    counts = result.get_counts(qc)
    plot_histogram(counts)

# Example usage
n_qubits = 3  # Number of qubits
marked_element = '110'  # Binary representation of the marked element
grover_algorithm(n_qubits, marked_element)

```

This above machine learning code defines a grover_algorithm function that takes the number of qubits n and the marked element as input. It creates a quantum circuit, applies Grover iterations, and measures the first n qubits. The example usage at the end demonstrates running the algorithm on a 3-qubit system with the marked element '110'.

# Main Code

Here's the main code that sets the number of qubits, the marked element, calls the grover_algorithm, visualizes the circuit, and plots the histogram:

```python
from qiskit import Aer, transpile
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

def main(num_qubits, marked_element):
    # Call grover_algorithm to create the Grover's algorithm circuit
    grover_algorithm(num_qubits, marked_element)

    # Visualize the circuit
    grover_circuit = QuantumCircuit(num_qubits + 1, num_qubits)
    initialize_s(grover_circuit, range(num_qubits + 1))
    oracle(grover_circuit, range(num_qubits), num_qubits, marked_element)
    diffusion(grover_circuit, range(num_qubits + 1))
    grover_circuit.measure(range(num_qubits), range(num_qubits))
    grover_circuit.draw(output='mpl')

    # Choose the Aer simulator backend
    simulator = Aer.get_backend('aer_simulator')

    # Run the circuit on the simulator
    compiled_circuit = transpile(grover_circuit, simulator)
    result = simulator.run(compiled_circuit).result()

    # Retrieve and plot the histogram of measurement outcomes
    counts = result.get_counts(grover_circuit)
    plot_histogram(counts)

    # Display the plots
    plt.show()

# Example usage
num_qubits = 3  # Number of qubits
marked_element = '110'  # Binary representation of the marked element
main(num_qubits, marked_element)
```

# Simulation and Visualization of Grover's algorithm circuit

```python
from qiskit import QuantumCircuit, Aer
from qiskit.visualization import plot_histogram

def initialize_s(qc, qubits):
    """Apply a Hadamard gate to create an equal superposition of states."""
    qc.h(qubits)
    qc.z(qubits)  # Apply Z-gates to convert |0 to |1

def oracle(qc, input_qubits, output_qubit):
    """Apply the oracle for the marked element."""
    qc.cx(input_qubits, output_qubit)

def diffusion(qc, qubits):
    """Apply the diffusion operator."""
    qc.h(qubits)
    qc.z(qubits)  # Apply Z-gates to convert |0 to |1
    qc.cz(qubits[-1], qubits[:-1])
    qc.h(qubits)

def grover_algorithm(n, marked_element):
    # Create quantum circuit with n qubits and one output qubit
    grover_circuit = QuantumCircuit(n + 1, n)

    # Initialize qubits
    initialize_s(grover_circuit, range(n + 1))

    # Define the number of Grover iterations for quadratic speedup
    num_iterations = int((3.14 / 4) * (2 ** (n / 2)))

    # Apply Grover iterations
    for _ in range(num_iterations):
        oracle(grover_circuit, range(n), n)
        diffusion(grover_circuit, range(n + 1))

    # Measure the first n qubits
    grover_circuit.measure(range(n), range(n))

    return grover_circuit

# Set the number of qubits and the marked element
num_qubits = 4
marked_element = 6  # This is the element we want to find

# Create the Grover's algorithm circuit
grover_circuit = grover_algorithm(num_qubits, marked_element)

# Visualize the circuit
grover_circuit.draw(output='mpl')

from qiskit import Aer

# Choose the Aer simulator
simulator = Aer.get_backend('aer_simulator')

# Run the circuit on the simulator
result = simulator.run(grover_circuit).result()

# Plot the histogram of measurement outcomes
counts = result.get_counts(grover_circuit)
plot_histogram(counts)
```

# Map classical inputs to a quantum problem

Mapping classical inputs to a quantum problem often involves encoding classical information into quantum states. Quantum circuits can be designed to perform specific transformations on qubits that represent classical data.

```python
from qiskit import QuantumCircuit, Aer, assemble
from qiskit.visualization import plot_histogram

def classical_to_quantum(circuit, classical_data, qubits):
    """Encode classical data into quantum states."""
    for index, value in enumerate(classical_data):
        if value == 1:
            circuit.x(qubits[index])

def quantum_to_classical(result, num_qubits):
    """Decode quantum states into classical data."""
    classical_data = [0] * num_qubits
    for index, count in enumerate(result.get_counts()):
        # Convert binary count to integer
        count_int = int(count, 2)
        # Check if the index is within the range
        if count_int < num_qubits:
            classical_data[count_int] = result.get_counts()[count]
    return classical_data

# Example classical input data
classical_input = [1, 0, 1]

# Number of qubits needed for encoding
num_qubits = len(classical_input)

# Create a quantum circuit
qc = QuantumCircuit(num_qubits, num_qubits)

# Encode classical input into quantum states
classical_to_quantum(qc, classical_input, range(num_qubits))

# Measure the qubits
qc.measure(range(num_qubits), range(num_qubits))

# Choose the Aer simulator
simulator = Aer.get_backend('aer_simulator')

# Assemble the circuit and execute it on the simulator
qobj = assemble(qc, backend=simulator)
result = simulator.run(qobj).result()

# Decode quantum states into classical data
decoded_classical_data = quantum_to_classical(result, num_qubits)

# Visualize the original classical input and the decoded result
print("Original Classical Input:", classical_input)
print("Decoded Classical Data:", decoded_classical_data)

# Plot the histogram of measurement outcomes
counts = result.get_counts(qc)
plot_histogram(counts)

```

In this python project notebook the main code creates a quantum circuit, encodes classical input into quantum states, measures the qubits, runs the circuit on a simulator, decodes the quantum states into classical data, and visualizes the results using a histogram.

classical_to_quantum function encodes classical input (binary data) into quantum states. It flips the qubits corresponding to the classical data that is set to 1.

quantum_to_classical function decodes quantum states into classical data. It converts the measurement outcomes into a classical binary representation.



## Grover's algorithm Background

Amplitude amplification is a general purpose quantum algorithm, or subroutine, that can be used to obtain a quadratic speedup over a handful of classical algorithms. [Grovers algorithm](https://arxiv.org/abs/quant-ph/9605043) was the first to demonstrate this speedup on unstructured search problems. Formulating a Grover's search problem requires an oracle function that marks one or more computational basis states as the states we are interested in finding, and an amplification circuit that increases the amplitude of marked states, consequently suppressing the remaining states.

Here, we demonstrate how to construct Grover oracles and use the `GroverOperator` from the Qiskit circuit library to easily set up a Grover's search instance. The runtime `Sampler` primitive allows seamless execution of Grover circuits, including automatic compilation, error suppression, and readout error mitigation techniques.

## Step 1: Map classical inputs to a quantum problem

Grover's algorithm requires an [oracle](https://learning.quantum-computing.ibm.com/course/fundamentals-of-quantum-algorithms/grovers-algorithm) that specifies one or more marked computational basis states, where "marked" means a state with a phase of -1.  A controlled-Z gate, or its multi-controlled generalization over $N$ qubits, marks the $2^{N}-1$ state (`'1'`\*$N$ bit-string).  Marking basis states with one or more `'0'` in the binary representation requires applying X-gates on the corresponding qubits before and after the controlled-Z gate; equivalent to having an open-control on that qubit.  In the following code, we define an oracle that does just that, marking one or more input basis states defined through their bit-string representation.  The `MCMT` gate is used to implement the multi-controlled Z-gate.


```python
def grover_oracle(marked_states):
    """Build a Grover oracle for multiple marked states

    Here we assume all input marked states have the same number of bits

    Parameters:
        marked_states (str or list): Marked states of oracle

    Returns:
        QuantumCircuit: Quantum circuit representing Grover oracle
    """
    if not isinstance(marked_states, list):
        marked_states = [marked_states]
    # Compute the number of qubits in circuit
    num_qubits = len(marked_states[0])

    qc = QuantumCircuit(num_qubits)
    # Mark each target state in the input list
    for target in marked_states:
        # Flip target bit-string to match Qiskit bit-ordering
        rev_target = target[::-1]
        # Find the indices of all the '0' elements in bit-string
        zero_inds = [ind for ind in range(num_qubits) if rev_target.startswith("0", ind)]
        # Add a multi-controlled Z-gate with pre- and post-applied X-gates (open-controls)
        # where the target bit-string has a '0' entry
        qc.x(zero_inds)
        qc.compose(MCMT(ZGate(), num_qubits - 1, 1), inplace=True)
        qc.x(zero_inds)
    return qc
```

### Specific Grover's instance

Now that we have the oracle function, we can define a specific instance of Grover search.  In this example we will mark two computational states out of the eight available in a three-qubit computational space:


```python
marked_states = ["011", "100"]

oracle = grover_oracle(marked_states)
oracle.draw(output="mpl", style="iqp")
```

### GroverOperator

The built-in Qiskit `GroverOperator` takes an oracle circuit and returns a circuit that is composed of the oracle circuit itself and a circuit that amplifies the states marked by the oracle.  Here, we `decompose` the circuit to see the gates within the operator:


```python
grover_op = GroverOperator(oracle)
grover_op.decompose().draw(output="mpl", style="iqp")
```

Repeated applications of this `grover_op` circuit amplify the marked states, making them the most probable bit-strings in the output distribution from the circuit.  There is an optimal number of such applications that is determined by the ratio of marked states to total number of possible computational states:


```python
optimal_num_iterations = math.floor(
    math.pi / 4 * math.sqrt(2**grover_op.num_qubits / len(marked_states))
)
```

### Full Grover circuit

A complete Grover experiment starts with a Hadamard gate on each qubit; creating an even superposition of all computational basis states, followed the Grover operator (`grover_op`) repeated the optimal number of times.  Here we make use of the `QuantumCircuit.power(INT)` method to repeatedly apply the Grover operator.


```python
qc = QuantumCircuit(grover_op.num_qubits)
# Create even superposition of all basis states
qc.h(range(grover_op.num_qubits))
# Apply Grover operator the optimal number of times
qc.compose(grover_op.power(optimal_num_iterations), inplace=True)
# Measure all qubits
qc.measure_all()
qc.draw(output="mpl", style="iqp")
```
